```python
def valida_mes():
    while True:
        try:
            mes = int(input("Digite o número do mês (1 a 12): "))
            if 1 <= mes <= 12:
                return mes
            else:
                print("Por favor, insira um número válido de mês (1 a 12).")
        except ValueError:
            print("Por favor, insira um número válido.")

def valida_temperatura():
    while True:
        try:
            temperatura = float(input("Digite a temperatura máxima para o mês em graus Celsius: "))
            if -90 <= temperatura <= 60:
                return temperatura
            else:
                print("Por favor, insira um valor de temperatura válido (-90 a 60 graus Celsius).")
        except ValueError:
            print("Por favor, insira um número válido para a temperatura.")

def nome_mes(numero):
    meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
    return meses[numero - 1]

def obter_dados_meteorologicos():
    dados_meteorologicos = []
    for i in range(12):
        print(f"\nDados para o mês {i+1}:")
        mes = valida_mes()
        temperatura = valida_temperatura()
        dados_meteorologicos.append((mes, temperatura))
    return dados_meteorologicos

def calcular_temperatura_media(dados_meteorologicos):
    soma_temperaturas = sum(t for m, t in dados_meteorologicos)
    return soma_temperaturas / 12

def calcular_meses_escaldantes(dados_meteorologicos):
    return sum(1 for m, t in dados_meteorologicos if t > 40)

def calcular_mes_mais_quente(dados_meteorologicos):
    mes, temperatura = max(dados_meteorologicos, key=lambda item: item[1])
    return nome_mes(mes), temperatura

def calcular_mes_menos_quente(dados_meteorologicos):
    mes, temperatura = min(dados_meteorologicos, key=lambda item: item[1])
    return nome_mes(mes), temperatura

# Coletando e analisando os dados
dados_meteorologicos = obter_dados_meteorologicos()
temperatura_media = calcular_temperatura_media(dados_meteorologicos)
meses_escaldantes = calcular_meses_escaldantes(dados_meteorologicos)
mes_mais_quente, temperatura_mais_quente = calcular_mes_mais_quente(dados_meteorologicos)
mes_menos_quente, temperatura_menos_quente = calcular_mes_menos_quente(dados_meteorologicos)

# Exibindo os resultados
print(f"\nTemperatura média máxima anual: {temperatura_media:.1f} graus Celsius")
print(f"Quantidade de meses escaldantes: {meses_escaldantes}")
print(f"Mês mais escaldante do ano: {mes_mais_quente} com {temperatura_mais_quente:.1f} graus Celsius")
print(f"Mês menos quente do ano: {mes_menos_quente} com {temperatura_menos_quente:.1f} graus Celsius")
```
